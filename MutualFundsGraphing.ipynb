```python
# !pip install yahoofinancials
# from yahoofinancials import YahooFinancials
# import datetime
# import matplotlib.pyplot as plt
# import pylab
# %matplotlib inline
# %pylab inline


def fundHist(ticker):
  plt.rcParams['figure.figsize'] = [25, 8]    #set graph size
  plt.xticks(rotation=90)   #set lables orientation

  now = datetime.datetime.today()
  tickerdata = YahooFinancials(ticker)
  pricehist = tickerdata.get_historical_price_data('2021-01-01', str(now.date()), 'daily')

  prices = {}
  for i in pricehist[ticker]['prices']:
    prices[i['formatted_date']] = round(i['close'],2)

  #if you want to see dates/prices
  #return prices

  #if you want to see graph
  return plt.plot(list(prices.keys()), list(prices.values()))
```
